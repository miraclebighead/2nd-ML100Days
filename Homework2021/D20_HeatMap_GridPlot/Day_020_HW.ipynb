```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣
matrix_10x10 = np.random.uniform(-1.0, 1.0, size=(10, 10))
print("10x10 矩陣 (數值分布 -1.0 ~ 1.0):\n", matrix_10x10)
print("-" * 30)


# 2. 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
matrix_1000x3_uniform = np.random.uniform(-1.0, 1.0, size=(1000, 3))
print("1000x3 矩陣 (均勻分布 -1.0 ~ 1.0):\n", matrix_1000x3_uniform)
print("-" * 30)


# 3. 建立一個 1000 x 3, 數值分布為常態分佈的矩陣
matrix_1000x3_normal = np.random.normal(loc=0.0, scale=1.0, size=(1000, 3))
print("1000x3 矩陣 (常態分布):\n", matrix_1000x3_normal)
```

```python
"""
Your Code Here
"""
matrix = matrix_10x10
plt.figure(figsize=(6,6))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap')
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = matrix_1000x3_uniform

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, height = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(sns.scatterplot , alpha = 0.2)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
